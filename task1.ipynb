```python
# Data Cleaning & Visualization Project

## About This Project
This project works on the famous Titanic dataset to demonstrate
data cleaning, preprocessing, and visualization techniques.

## Dataset
- **Source:** Titanic passenger data
- **Rows:** 891 passengers
- **Columns:** 12 features

## Tools Used
- **Pandas** - Data cleaning and manipulation
- **Matplotlib** - Data visualization
- **Seaborn** - Statistical visualizations

## Steps Followed
1. Loaded the dataset
2. Explored the data (shape, info, statistics)
3. Found and handled missing values
4. Removed duplicate rows
5. Created visualizations
6. Drew conclusions from charts

## Key Findings
- Females survived more than males
- 1st class passengers had better survival rates
- Most passengers were aged between 20 to 40 years
```

```python
import pandas as pd

# Load Titanic dataset directly from the internet
df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")

print("Dataset loaded!")
print(df.shape)  # shows rows and columns
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
print("Age fixed!")
```

```python
df.isnull().sum()
```

```python
df['Embarked'].fillna(df['Embarked'].mode()[0],inplace=True)
print("Embarked fixed!")
```

```python
df.drop(columns=['Cabin'], inplace=True)
print("Cabin removed!")
```

```python
print("Duplicates before:", df.duplicated().sum())
df.drop_duplicates(inplace=True)
print("Duplicates after:", df.duplicated().sum())
```

```python
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.countplot(x='Survived', data=df)
plt.title('Survival Count (0 = Died, 1 = Survived)')
plt.show()
```

```python
sns.countplot(x='Survived', hue='Sex', data=df)
plt.title('Survival by Gender')
plt.show()
```

```python
sns.histplot(df['Age'], bins=30, kde=True)
plt.title('Age Distribution of Passengers')
plt.show()
```

```python
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title('Survival by Passenger Class')
plt.show()
```

```python
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df)
plt.title('Age vs Fare by Survival')
plt.show()
```
